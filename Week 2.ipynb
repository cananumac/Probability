Para Atma Deneyi

```python
def my_f(n):
    h_number=0     
    t=[]            #boş bir liste oluşturduk
    for i in range(n):              # parayı n defa attık.
        outcome=np.random.choice(['H','T'])
        t.append(outcome)              #çıkan değerler listeye ekledik.
        #print(i,"th expeirment: ",outcome)  #sonuçları ekrana yazdırdık.
        if outcome=='H':             
            h_number=h_number+1     #eğer head gelirse h_number değişkenini bir arttır.
    return h_number/n   #head gelme olasılığını geri döndür.
```

```python
my_f(100)
```

Zar Atma Deneyi

```python
def my_f_for_die(n):
    number_of_1=0  
    number_of_4=0 
    t=[]            #boş bir liste oluşturduk
    for i in range(n):              #zarı n defa attık.
        outcome=np.random.choice([1,2,3,4,5,6])
        t.append(outcome)              #çıkan değerleri listeye ekledik.
        #print(i,"th expeirment: ",outcome)  
        if outcome==1:               
            number_of_1=number_of_1+1   #eğer 1 gelirse number_of_1 değişkenini bir arttır.
            
        if outcome==4:             
            number_of_4=number_of_4+1   #eğer 4 gelirse number_of_4 değişkenini bir arttır.         
            
    return number_of_1/n,number_of_4/n #bir ve dört gelme olasılığını geri döndür.
```

```python
my_f_for_die(100)
```

Bir şapkanın içinde 3 adet kart vardır.İlk kartın der iki tarafı da siyah, ikinci kartın her iki tarafı da beyaz ve son kartın bir tarafı siyah diğer tarafı beyazdır.

```python
import random
```

```python
def my_f_cards(p=False):
    b_number=0
    w_number=0
    c_1=['black','black']
    c_2=['white','white']
    c_3=['black','white']
    cards=[c_1,c_2,c_3]
    select_card=random.choice(cards)  #3 karttan rastgele bir tanesini seçiyoruz.
    if p:
        print(select_card,end=" ") 
    select_face=random.choice(select_card) #seçilen kartın bir yüzünü seçiyoruz.
    select_card.remove(select_face)
    unseen_face=select_card[0] 
    if p:
        print("Görünen yüz: ",select_face," Bilinmeyen yüz: ",unseen_face,end=" ")
    
    if select_face=='black':   #seçilen yüz siyah ise b_number değişkenini bir arttırıyoruz.
        b_number=b_number+1
        if p:
            print("   + ")
        if unseen_face=='white':  #görünmeyen yüz beyaz ise w_number değerini bir arttırıyoruz. 
            w_number=w_number+1
        
    else:
        if p:
            print()
    return b_number,w_number  
    
```

```python
def my_f_cards2(n):      # n defa kart çekerek b_number ve w_number değerlerini hesaplayıp geri döndürüyoruz.
    b=0
    w=0
    for i in range(n):
        b=b+ my_f_cards()[0]
        w=w+my_f_cards()[1]
    return n,b,w
```

```python
def my_f_cards3(n):   #Geri döndürülen değerlerin olasılık değerlerini hesaplıyoruz.
    seen_face_number=0
    unseen_face_number=0

    for i in range(n):
        a,b,c=my_f_cards2(n)
        seen_face_number=seen_face_number+b
        unseen_face_number=unseen_face_number+c
    return int(seen_face_number/n),int(unseen_face_number/n)
```
